```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact
from IPython.display import display

# Assuming you have already loaded the dataset
app_details = pd.read_csv('/content/Screentime_APPdetails.csv')
app_details['Date'] = pd.to_datetime(app_details['Date'], errors='coerce')

# Rest of your code...
def plot_times_opened(app):
    plt.figure(figsize=(10, 6))
    filtered_data = app_details[app_details['App'] == app]
    sb.lineplot(x=filtered_data['Date'], y=filtered_data['Times opened'])
    plt.title(f'Number of times {app} was opened over time')
    plt.show()

def plot_notifications(app):
    plt.figure(figsize=(10, 6))
    filtered_data = app_details[app_details['App'] == app]
    sb.lineplot(x=filtered_data['Date'], y=filtered_data['Notifications'])
    plt.title(f'Number of notifications for {app} over time')
    plt.show()

# Get unique app names
app_names = app_details['App'].unique()

# Create interactive widgets
app_selector = widgets.Dropdown(options=app_names, description='Select App:')
plot_type_selector = widgets.Dropdown(options=['Times Opened', 'Notifications'], description='Select Plot Type:')

# Define the interactive function
def interactive_plot(app, plot_type):
    if plot_type == 'Times Opened':
        plot_times_opened(app)
    elif plot_type == 'Notifications':
        plot_notifications(app)

# Set up the interactive widget
interact(interactive_plot, app=app_selector, plot_type=plot_type_selector)
```

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact
from IPython.display import display

# Assuming you have already loaded the dataset
app_details = pd.read_csv('/content/Screentime_APPdetails.csv')
app_details['Date'] = pd.to_datetime(app_details['Date'], errors='coerce')

# ... (the rest of your code)


# Define a function to create plots based on user input
def create_plots(app_details, plot_type='countplot', app=None):
    plt.figure(figsize=(10, 6))

    if plot_type == 'countplot':
        sb.countplot(x=app_details['App'])
        plt.xticks(rotation=120)
        plt.title('Number of records for each app')
    elif plot_type == 'lineplot':
        sb.lineplot(x=app_details['Date'], y=app_details['Times opened'], hue=app_details['App'])
        plt.xticks(rotation=120)
        plt.title('Number of times each app was opened over time')

    plt.show()

# Create interactive widgets
plot_type_selector = widgets.Dropdown(options=['countplot', 'lineplot'], description='Select Plot Type:')
app_selector = widgets.Dropdown(options=['All'] + list(app_details['App'].unique()), description='Select App:')

# Set up the interactive function
def interactive_plot(plot_type, app):
    if app == 'All':
        create_plots(app_details, plot_type=plot_type)
    else:
        filtered_data = app_details[app_details['App'] == app]
        create_plots(filtered_data, plot_type=plot_type)

# Set up the interactive widget
interact(interactive_plot, plot_type=plot_type_selector, app=app_selector)
```

```python
import numpy as np
import pandas as pd
import seaborn as sb
import plotly.express as ex
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interactive

# Load the dataset
app_details = pd.read_csv('/content/Screentime_APPdetails.csv')
app_details['Date'] = pd.to_datetime(app_details['Date'], errors='coerce')  # Add 'errors' parameter

# Define a function for interactive plotting
def plot_timeseries(start_date, end_date):
    selected_data = app_details[(app_details['Date'] >= start_date) & (app_details['Date'] <= end_date)]

    # Plot lineplot
    plt.figure(figsize=(15, 8))
    plt.subplot(2, 1, 1)
    sb.lineplot(x='Date', y='Times opened', hue='App', data=selected_data)
    plt.title('Number of times each app opened during the selected time range')

    # Plot boxplot
    plt.subplot(2, 1, 2)
    sb.boxplot(x='App', y='Times opened', data=selected_data)
    plt.title('Boxplot of the number of times each app opened during the selected time range')

    plt.tight_layout()
    plt.show()

    # Plot lineplot
    plt.figure(figsize=(15, 8))
    plt.subplot(2, 1, 1)
    sb.lineplot(x='Date', y='Notifications', hue='App', data=selected_data)
    plt.title('Number of notifications during the selected time range for each app')

    # Plot boxplot
    plt.subplot(2, 1, 2)
    sb.boxplot(x='App', y='Notifications', data=selected_data)
    plt.title('Boxplot of the number of notifications during the selected time range for each app')

    plt.tight_layout()
    plt.show()

# Create interactive widgets
start_date_widget = widgets.DatePicker(value=min(app_details['Date']), description='Start Date')
end_date_widget = widgets.DatePicker(value=max(app_details['Date']), description='End Date')

# Create interactive plot
interactive_plot = interactive(plot_timeseries, start_date=start_date_widget, end_date=end_date_widget)
output = interactive_plot.children[-1]
output.layout.height = '800px'
interactive_plot
```

```python
import pandas as pd
from scipy.stats import skew, kurtosis

# Read the CSV file into a DataFrame
file_path = '/content/Screentime_APPdetails.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
print("Data Preview:")
print(data.head())

# Select only numeric columns
numeric_columns = data.select_dtypes(include=['number'])

# Calculate skewness and kurtosis for each numeric column
skewness = numeric_columns.apply(skew)
#Positive Skewness (Right-skewed): If the skewness is greater than 0.The mean is typically greater than the median in a right-skewed distribution.
#Negative Skewness (Left-skewed): If the skewness is less than 0.The mean is typically less than the median in a left-skewed distribution.
#If the skewness is close to 0, the distribution is approximately symmetrical.
kurt = numeric_columns.apply(kurtosis)
#kurtosis provides insights into the shape of the tails of a distribution, helping to identify whether a dataset has more or fewer extreme values than a normal distribution.

# Display skewness and kurtosis for each numeric column
print("\nSkewness:")
print(skewness)

print("\nKurtosis:")
print(kurt)
```

```python
import pandas as pd

# Assuming your CSV file is named 'your_file.csv'
file_path = '/content/Screentime_APPdetails.csv'

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path)

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce', infer_datetime_format=True)

# Group by 'Date' and calculate the sum for each group
result = df.groupby(['Date', 'App']).agg({
    'Usage': 'sum',
    'Notifications': 'sum',
    'Times opened': 'sum'
}).reset_index()

# Sum the values for each date
total_result = result.groupby('Date').agg({
    'Usage': 'sum',
    'Notifications': 'sum',
    'Times opened': 'sum'
}).reset_index()

# Print the result
print(total_result)
```

```python
import pandas as pd

# Assuming your DataFrame is named 'total_result' from the previous code
# If not, replace 'total_result' with the actual DataFrame name

# Add new columns for 'Usage Decrease', 'Notifications Decrease', and 'Times opened Decrease'
total_result['Usage Decrease'] = 0
total_result['Notifications Decrease'] = 0
total_result['Times opened Decrease'] = 0

# Identify rows where each column individually is greater than 100 and calculate the decrease
usage_mask = total_result['Usage'] > 100
notifications_mask = total_result['Notifications'] > 100
times_opened_mask = total_result['Times opened'] > 100

total_result.loc[usage_mask, 'Usage Decrease'] = total_result['Usage'] - 100
total_result.loc[notifications_mask, 'Notifications Decrease'] = total_result['Notifications'] - 100
total_result.loc[times_opened_mask, 'Times opened Decrease'] = total_result['Times opened'] - 100

# Print the updated DataFrame
print(total_result)
```

```python
import pandas as pd
import numpy as np
import random

# Define the list of city names
city_names = ["New York", "Los Angeles", "Chicago", "Houston", "Philadelphia", "Phoenix", "San Antonio", "San Diego", "Dallas", "San Jose"]

# Define the list of professions
professions = ["Software Engineer", "Data Scientist", "Marketing Manager", "Accountant", "Salesperson", "Teacher", "Nurse", "Doctor", "Lawyer", "Engineer"]

# Create a list to store the fake data
data = []

# Generate 500 entries of fake data
for i in range(500):
    # Generate random values for each column
    name = random.choice(names)
    profession = random.choice(professions)
    city = random.choice(city_names)
    data_usage = random.randint(100, 300)
    notifications = random.randint(5, 50)
    time_of_usage = random.randint(1, 24)

    # Create a dictionary to store the data
    entry = {
        "Name": name,
        "Profession": profession,
        "City": city,
        "Data Usage": data_usage,
        "Notifications": notifications,
        "Time of Usage": time_of_usage
    }

    # Add the dictionary to the data list
    data.append(entry)

# Create a pandas DataFrame from the list of dictionaries
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv("fake_data.csv", index=False)
```
